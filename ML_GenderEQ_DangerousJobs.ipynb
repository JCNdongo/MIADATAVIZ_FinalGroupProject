### <center> Prediction Based on Aggregated Data for 141 Countries </center> ##

* Dependent Variable Name: dangerous_job
* Dependent Variable Meaning: A woman can work in a job deemed dangerous in the same way as a man (1=yes; 0=no)



```python
# Import dependencies
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
```

## <center> Split the Data into Training and Testing Sets </center> ##
### Step 1: Read the lending_data.csv data from the Resources folder into a Pandas DataFrame. ###

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
file_path = Path("CleanData/COVID_FinalProject.csv")
covid_df = pd.read_csv(file_path)

# Review the DataFrame
covid_df.head()
```

```python
covid_new_df = covid_df.drop(columns=['Country_Code', 'Country_Name', 'Continent', 'Income', 'Agency_Name', 'Agency_Acronym'])
covid_new_df.head()
```

### Step 2: Create the labels set (y) from the “divorce_choice” column, and then create the features (X) DataFrame from the remaining columns. ###

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = covid_new_df['dangerous_job']

# Separate the X variable, the features
X = covid_new_df.drop(columns=['dangerous_job'])
```

```python
# Review the y variable Series
y.head()
```

```python
# Review the X variable DataFrame
X.head()
```

### Step 3: Check the balance of the labels variable (y) by using the value_counts function. ###

```python
# Check the balance of our target values
y.value_counts()
```

### Step 4: Split the data into training and testing datasets by using train_test_split. ###

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

## <center> Create a Logistic Regression Model with the Original Data </center> ##
### Step 1: Fit a logistic regression model by using the training data (X_train and y_train). ###

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
logit_regression = LogisticRegression(solver='lbfgs', random_state=1)

# Fit the model using training data
logit_model = logit_regression.fit(X_train, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (X_test) and the fitted model. ###

```python
# Make a prediction using the testing data
dangerous_job_predictions = logit_model.predict(X_test)
```

### Step 3: Evaluate the model’s performance by doing the following: ###
* Calculate the accuracy score of the model. 

* Generate a confusion matrix. 

* Print the classification report. 

```python
# Print the balanced_accuracy score of the model
#Import the accuracy_score module from sklearn
from sklearn.metrics import accuracy_score

print(f"Balanced Accuracy Score: {balanced_accuracy_score(y_test, dangerous_job_predictions)}")
```

```python
# Generate a confusion matrix for the model
from sklearn.metrics import confusion_matrix
testing_model_matrix = confusion_matrix(y_test, dangerous_job_predictions)
print(testing_model_matrix)
```

```python
# Print the classification report for the model
dangerous_job_report = classification_report(y_test, dangerous_job_predictions)
print(dangerous_job_report)
```

## <center> Results: 85% Accuracy Rate </center> ##

* Our results predicted women's ability to get similarly dangerous jobs as men worldwide following COVID with a 85% accuracy rate. 

* This high accuracy rate reflects that the variables taken into consideration to produce this model were adequate. 

* This model aggregated 141 countries across the world (different continents and economic indicators). Although the model is robust, it could have been improved if it would have focused on specific countries, regions, or continents with similar socio-economic indicators.
